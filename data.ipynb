```python
from google.cloud import storage
import os

def download_blob(bucket_name, source_blob_name, destination_file_name):
    """Downloads a blob from the bucket."""
    # bucket_name = "your-bucket-name"
    # source_blob_name = "storage-object-name"
    # destination_file_name = "local/path/to/file"

    storage_client = storage.Client()

    bucket = storage_client.bucket(bucket_name)

    # Construct a client side representation of a blob.
    # Note `Bucket.blob` differs from `Bucket.get_blob` as it doesn't retrieve
    # any content from Google Cloud Storage. As we don't need additional data,
    # using `Bucket.blob` is preferred here.
    blob = bucket.blob(source_blob_name)
    blob.download_to_filename(destination_file_name)

    print(
        "Blob {} downloaded to {}.".format(
            source_blob_name, destination_file_name
        )
    )

# set credential path
# https://cloud.google.com/storage/docs/reference/libraries
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="D:/myfiles/brandies/Spring2021/COSI-149B/teset-918f5cd7a22a.json"
```

create train folders

```python
import shutil,os
```

```python
root = os.getcwd()
root
if os.path.exists(root + "/project1/train"):
    shutil.rmtree(root + "/project1/train")
if os.path.exists(root + "/project1/test"):
    shutil.rmtree(root + "/project1/test")
if os.path.exists(root + "/project1/mean_std_value_train.pkl"):
    os.remove(root + "/project1/mean_std_value_train.pkl")
if os.path.exists(root + "/project1/mean_std_value_test.pkl"):
    os.remove(root + "/project1/mean_std_value_test.pkl")
os.mkdir(root + "/project1/train")
os.mkdir(root + "/project1/train/true")
os.mkdir(root + "/project1/train/false")
os.mkdir(root + "/project1/test")
os.mkdir(root + "/project1/test/true")
os.mkdir(root + "/project1/test/false")

path_raw="project1/raw/"
path_train_true="project1/train/true/"
path_train_false="project1/train/false/"
path_test_true="project1/test/true/"
path_test_false="project1/test/false/"
```

```python
# I use first 1600 files to test
df = doc[0:1600].copy()
#shuffle data
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
# train using 90% data, rest 10% for test
size = len(df)
div = int(size/10*9)
#train
for index, row in df[0:div].iterrows():
    filename = row["filename"][6:]
    if row["class"] > 1:
        shutil.copy(path_raw + filename, path_train_true + filename)
    else:
        shutil.copy(path_raw + filename, path_train_false + filename)
#test
for index, row in df[div:].iterrows():
    filename = row["filename"][6:]
    if row["class"] > 1:
        shutil.copy(path_raw + filename, path_test_true + filename)
    else:
        shutil.copy(path_raw + filename, path_test_false + filename)
```
